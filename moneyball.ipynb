```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
%matplotlib inline
```

```python
batting_stats = pd.read_csv('data/Batting.csv', sep=',', low_memory=False)
```

```python
salary_stats = pd.read_csv('data/Salaries.csv', sep=',', low_memory=False)
```

```python
appearances_stats = pd.read_csv('data/Appearances.csv', sep=',', low_memory=False)
```

```python
bio_info = pd.read_csv('data/Master.csv', sep=',', low_memory=False)
```

```python
bio = bio_info[['playerID', 'nameFirst', 'nameLast']]
```

```python
batting = batting_stats[['playerID', 'yearID', 'teamID', 'AB', 'H', 'BB', 'HBP', 'SF']]
```

```python
salary = salary_stats[['playerID', 'yearID', 'teamID', 'salary']]
```

```python
position = appearances_stats[['playerID', 'yearID', 'teamID', 'G_p', 'G_c', 'G_1b', 'G_2b', 'G_3b', 'G_ss', 'G_lf', 'G_cf', 'G_rf', 'G_of', 'G_dh', 'G_all']]
```

```python
players = pd.merge(pd.merge(pd.merge(bio,batting), position), salary)
```

```python
players.head()
```

```python
# players = players[players['G_all'] > 30]
```

```python
players = players.set_index(['playerID'])
```

```python
players.head()
```

```python
players['OBP'] = (players['H']+players['BB']+players['HBP'])/(players['AB']+players['BB']+players['HBP']+players['SF'])
```

```python
players['OBP_$_ratio'] = (players['OBP']/players['salary'])*1000000
```

## calculates On Base "Percentage" from recorded stats. 
## then calculates a ratio to the players salary which is multiplied by 1,000,000 to relate to players making more than $1,000,000. 
## this also makes the number more readable.



```python
play_more20 = players['G_all'] >= 20
```

```python
play_more20 = players[play_more20]
```

## takes away players with less than 20 games played per year

```python
no_pitchers = play_more20['G_p'] <= 1

```

```python
no_pitchers = play_more20[no_pitchers]
```

## takes away pitchers, who cares about their batting, their job is to get outs

```python
players_filtered = no_pitchers.copy()
```

```python
sorted_by_ratio = players_filtered.sort(['OBP_$_ratio'], ascending=False)
```

## sorts with highest OBP ratio on top

```python
sorted_by_ratio.head()
```

```python
def one_year(year):
    return sorted_by_ratio[sorted_by_ratio.yearID == year]
```
